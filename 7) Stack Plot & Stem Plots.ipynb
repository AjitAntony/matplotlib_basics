# Stack Plot

view composition(pie chart) and time trends(line chart ) in same chart 

trends in variables over a specific period of time.

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
trade_data=pd.read_csv("US_trade.csv")
trade_data
#trade between us and chine was 2.1% of all the us trade 
```

```python
trade_data.columns
```

```python
# take arrays and stack vertically
import numpy as np
data_vstack=np.vstack([
    trade_data["China"],
     trade_data["Japan"],
     trade_data["Mexico"],
     trade_data["Canada"],
     trade_data["Germany"],
     trade_data["France"],
    trade_data["India"],
    trade_data["UK"],
    trade_data["Taiwan"],
]
)
data_vstack
```

```python
plt.figure(figsize=(12,6))
plt.stackplot(trade_data["Year"],data_vstack,labels=[ 'China', 'Japan', 'Mexico', 'Canada', 'Germany', 'France','India', 'UK', 'Taiwan'])
plt.legend()
```

```python
# trade btw japan and us got shrunk over year while its opoiste case for china 
# germany trad was very stable and did not go high/low
```

# Stem plot 

 stem plot plots vertical lines at each x location from the baseline to y, and places a marker there.
 

```python
trade_data_diff=trade_data.copy()
```

```python
data_columns=[ 'China', 'Japan', 'Mexico', 'Canada', 'Germany', 'France','India', 'UK', 'Taiwan']
```

```python
trade_data_diff[data_columns]=trade_data_diff[data_columns].diff()  #compute difference between current row and previous row 
trade_data_diff

# trade between us and japan fell by 1.2% in 1996 when compared to 1995
# trade between us and gremany grew  by 0.5% in 1998 when compared to 1997

```

```python
plt.figure(figsize=(12,6))
plt.stem(trade_data_diff["Year"],trade_data_diff["China"])
```

```python
# from 1990-2005 trade was smooth did not fell down drastically
# from 2005 the trade started  to decline and rise up and this pattern continued 
```

```python
fig,ax=plt.subplots(9,1,figsize=(30,40))

ax[0].stem(trade_data_diff["Year"],trade_data_diff["China"])
ax[1].stem(trade_data_diff["Year"],trade_data_diff["Japan"])
ax[2].stem(trade_data_diff["Year"],trade_data_diff["Mexico"])
ax[3].stem(trade_data_diff["Year"],trade_data_diff["Canada"])
ax[4].stem(trade_data_diff["Year"],trade_data_diff["Germany"])
ax[5].stem(trade_data_diff["Year"],trade_data_diff["France"])
ax[6].stem(trade_data_diff["Year"],trade_data_diff["India"])
ax[7].stem(trade_data_diff["Year"],trade_data_diff["UK"])
ax[8].stem(trade_data_diff["Year"],trade_data_diff["Taiwan"])
```
