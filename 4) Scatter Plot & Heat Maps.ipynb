# Scatter Plot

    Identify relationships between features in data weak/strong correlation 
    Helps to find outliers and spreadness of data 



# load Libraries

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Simple scatter plot

```python
no_of_adv=[10,15,20,25,30,40,50]
sales=[100,130,200,270,300,400,1000]

fig,ax=plt.subplots()
ax.scatter(no_of_adv,sales)
ax.set_xlabel('Ads')
ax.set_ylabel('Sales')
```

# Scatter plot 'Time on App' vs 'Yearly Amount Spent'

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.scatter(ecommerce_data[ 'Time on App'],ecommerce_data['Yearly Amount Spent'])
ax.set_xlabel('Time on App')
ax.set_ylabel('Yearly Amount Spent')
```

# Customize 



```python
fig,ax=plt.subplots(figsize=(12,6))
ax.scatter(ecommerce_data[ 'Time on App'],ecommerce_data['Yearly Amount Spent'],s=50,c="red",marker="X")
ax.set_xlabel('Time on App')
ax.set_ylabel('Yearly Amount Spent')
```

### Relation between 2 variables along with 3rd axis 

```python
fig,ax=plt.subplots(figsize=(12,6))
ax_res=ax.scatter(ecommerce_data[ 'Time on App'],ecommerce_data['Yearly Amount Spent'],s=50,c=ecommerce_data['Length of Membership'],
           marker="X",cmap="Reds")
# cmap : https://matplotlib.org/stable/tutorials/colors/colormaps.html
ax.set_xlabel('Time on App')
ax.set_ylabel('Yearly Amount Spent')
fig.colorbar(ax_res,label="Length of Membership",aspect=50)
```

# visualize 2 scatter points side by side 

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))

ax[0].scatter(ecommerce_data[ 'Time on Website'],
              ecommerce_data['Yearly Amount Spent'],
              s=50,
              c=ecommerce_data['Length of Membership'],
              marker="X",cmap="Reds")
# cmap : https://matplotlib.org/stable/tutorials/colors/colormaps.html
ax[0].set_xlabel('Time on Website')
ax[0].set_ylabel('Yearly Amount Spent')



ax[1].scatter(ecommerce_data[ 'Time on App'],
              ecommerce_data['Yearly Amount Spent'],
              s=50,
              c=ecommerce_data['Length of Membership'],
              marker="X",cmap="Reds")
# cmap : https://matplotlib.org/stable/tutorials/colors/colormaps.html
ax[1].set_xlabel('Time on App')
ax[1].set_ylabel('Yearly Amount Spent')
```

# Heat Maps

    Represent data using color codes .
    Graphical representation of data where each value of a matrix is represented as a color.
    
    

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

farm_data=pd.DataFrame(harvest,columns=farmers,index=vegetables)
```

# Harvest of local farmers in tons per year

```python
farm_data
```

```python
fig, ax = plt.subplots()
im = ax.imshow(harvest,cmap="Reds")
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)   

plt.setp(ax.get_xticklabels(),rotation=45) 
fig.colorbar(im)
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        #print(i,j)
        text = ax.text(j, i, harvest[i, j])
    
```

# Correlation matrix : Find  relationship for large number of varibles taken 2 at time

    +1 = postiively correlated

    -1 = negative correlated 

```python
ecommer_corr=ecommerce_data.corr()
ecommer_corr
```

```python
# convert to numpy 
ecommer_corr_numpy=ecommer_corr.to_numpy()
fig,ax=plt.subplots(figsize=(20,5))
#disaplys data as image , data must be a matrix and it will be rendered as colored image
im=ax.imshow(ecommer_corr_numpy,cmap="Reds")
```

```python
fig,ax=plt.subplots(figsize=(20,5))

#disaplys data as image , data must be a matrix and it will be rendered as colored image
im=ax.imshow(ecommer_corr_numpy,cmap="Reds")


features = list(ecommer_corr.columns)

ax.set_xticks([0,1,2,3,4])
ax.set_yticks([0,1,2,3,4])

ax.set_xticklabels(features)
ax.set_yticklabels(features)

#roate xticklabels so that ther wont overlap 
plt.setp(ax.get_xticklabels(),rotation=45) 

#plot the corr values under each cell 
for i in range(len(features)):
    for j in range(len(features)):
        text=ax.text(j,i,"%.2f"% float (ecommer_corr_numpy[i,j]))
```
