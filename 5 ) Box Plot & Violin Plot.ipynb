# box-and-whisker plot

    A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. 
    
    


# Import Libraries 

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# customizaton

```python

fig,ax=plt.subplots(figsize=(12,6))

ax.boxplot([customer_churn["Subscription  Length"],customer_churn["Frequency of use"],customer_churn["Frequency of SMS"]],
                      showfliers=False,
          showmeans=True,vert=0)
ax.set_xticklabels(['Subscription  Length','Frequency of use','Frequency of SMS'])
```

# Violine Plot 

    Violin plots is combination of  histograms and box plots and in addition to showing the summary statistics it also shows the distribution of data in particular column/feature.
    
    It shows original & mirrored version of Histogram  
    Box plot shows potential oultiers in data along with IQR
    
    Violin plot shows the KDE  to compute an empirical distribution of the sample

# simple violine plot 

```python

fig,ax=plt.subplots()
 
ax.violinplot(customer_churn["Subscription  Length"] ,showmedians=True
)
```

 # violine vs box vs histograms

```python

fig,ax=plt.subplots(1,3,figsize=(20,5))
#pass and array of sequence of vectot if you just pass customer_churn_mini it wont diaply the plot 
ax[0].boxplot(customer_churn["Subscription  Length"])
 

ax[1].violinplot(customer_churn["Subscription  Length"] ,showmedians=True,showmeans=True
)
             
ax[2].hist(customer_churn["Subscription  Length"],orientation="horizontal")
 
 
```
