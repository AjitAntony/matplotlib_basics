# Interfaces

    There are two types of interfaces in Matplotlib for visualization 
    1.Object Oriented interface
        when there is a need to draw Multiple plots on the canvas or to draw  more complex graphs with customization  .
    
    2.state-based interface (using pyplot )
        The pyplot interface is used for simple plot generation and doesn't really scale well when there is a need to make multiple plots or when we have to make plots that require a lot of customization.

 

# Matplotlib Object Hierarchy

        1.Figure object : can be thought of as a canvas, upon which different elements(artists) act to create the final graphic image .
        2.Artists : All of the items labelled in blue are Artists
        
        3.Axes object : An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects 
    
        4.Axis Object : Each Axes object will contain two or more Axis objects.  It will set the data scale & limits and also contains the ticks and ticks labels.

# Part B:  Object-Oriented Interface

    state based interface : plt.figure()
    
    Object-Oriented Interface : plt.subplots 
    
    
    subplots - represent multiple plots in single figure 

    plt.subplots() takes in 3 arguments that decides the figure layout(layout is organized as rows,columns,index ) by default rows=1,columns=1

    we can group mulitple axes in same figure and have individual control on them 

    Given the number of rows and columns, it returns a tuple (fig, ax), giving a single figure fig with an array of axes ax.

## Figure 1 : Empty figure plot

```python
# Note 1stline of code uses pyplots plt alias we are mixing pyplot and object oriented api 
fig,ax=plt.subplots()
# create figure and group of axes that can exist in in single figure
ax
```

## Figure 2 : Figure with  Plot  added single list of values to the plot property

```python
age=[10,20,30,40,50]
height_in_feet=[5.2,5.15,5.2,5.2,5.5]
weight_in_kg=[30,60,90,120,150]
```

```python
# create figure and group of axes that can exist in in single figure
fig,ax=plt.subplots()
# Create a figure containing a single axes by default
ax.plot(age,"bo-")
```

 ## Figure 3 : Age vs Height 

```python
# Create a figure containing a single axes by default
fig,ax=plt.subplots()
# plot Age vs Height
ax.plot(age,height_in_feet,"bo-")
ax.set_xlabel("age")
ax.set_ylabel("height(ft)")
```

 ## Figure 4 : Age vs Weight 

```python
# Create a figure containing a single axes by default
fig,ax=plt.subplots()
# plot Age vs Height
ax.plot(age,weight_in_kg,"bo-")
ax.set_xlabel("age")
ax.set_ylabel("weight(kg)")
```

   ## A) Stacking subplots in one direction

When stacking in one direction only, the returned axs is a 1D numpy array containing the list of created Axes.

```python
fig,ax=plt.subplots(1,2)
#ax,ax[0],ax[1]
#axes
ax[0].plot(age,height_in_feet,"bo-")
ax[0].set_xlabel("age")
ax[0].set_ylabel("height(ft)")


ax[1].plot(age,weight_in_kg,"bo-")
ax[1].set_xlabel("age")
ax[1].set_ylabel("weight(kg)")
```

   ## B) Stacking subplots in one direction

When stacking in one direction only, the returned axs is a 1D numpy array containing the list of created Axes.Stacking subplots in two directions

```python
fig,ax=plt.subplots(2,1)
ax,ax[0],ax[1]

ax[0].plot(age,height_in_feet,"bo-")
ax[0].set_xlabel("age")
ax[0].set_ylabel("height(ft)")


ax[1].plot(age,weight_in_kg,"bo-")
ax[1].set_xlabel("age")
ax[1].set_ylabel("weight(kg)")
```



When stacking in one direction only, the returned axs is a 1D numpy array containing the list of created Axes.

   ## C)  create multiple lines in the same line chart with different y-axes for each line.

```python
#we need 2 axes object with one common x axis and 2 independent yaxis 
# There will be Two plots on the same axes with different left and right scales.

#axes-ax1
fig,ax1=plt.subplots()
ax1.plot(age,height_in_feet,"bo-")
ax1.set_xlabel("age")
ax1.set_ylabel("height(ft)")
#Create a twin Axes sharing the xaxis of above first axes but will have its own independent y-axis 

#axes - ax2
ax2=ax1.twinx()
ax2.plot(age,weight_in_kg,"go-")
ax2.set_ylabel("weight(kg)")
```
