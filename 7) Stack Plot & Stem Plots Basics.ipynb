# Stack Plot

view composition(pie chart) and time trends(line chart ) in same chart 

trends in variables over a specific period of time.

# Import library 

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Load Dataset

```python
trade_data=pd.read_csv("US_trade.csv")
trade_data
```

# Vertical stack array

```python
data_vstack=np.vstack([
    trade_data["China"],
     trade_data["Japan"],
     trade_data["Mexico"],
     trade_data["Canada"],
     trade_data["Germany"],
     trade_data["France"],
    trade_data["India"],
    trade_data["UK"],
    trade_data["Taiwan"],
]
)
data_vstack
```

# stack plot for trades

```python
plt.figure(figsize=(12,6))
plt.stackplot(trade_data["Year"],data_vstack,
              labels=[ 'China', 'Japan', 'Mexico', 'Canada', 'Germany', 'France','India', 'UK', 'Taiwan'])
plt.legend()
```

# Stem plot 

 stem plot plots vertical lines at each x location ( baseline) to y coordinate, and places a marker at tip .
 

# Simple Stem plots

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))
print(x,y)
plt.stem(x, y)
```

# Copy the dataframe/dataset

```python
trade_data_diff=trade_data.copy()
```

# Column Names 

```python
data_columns=list(trade_data_diff.columns)[1:]
data_columns
```

# Compute the trade difference 
compute difference between current row and previous row 

```python
trade_data_diff
```

```python
#compute difference between current row and previous row 
trade_data_diff[data_columns]=trade_data_diff[data_columns].diff() 
trade_data_diff
```
